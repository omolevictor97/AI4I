```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression, Lasso, Ridge
```

```python
X = df.drop("Air temperature [K]", axis=1)
y = df["Air temperature [K]"]

from sklearn.feature_selection import VarianceThreshold

varia_thresh = VarianceThreshold(threshold=0.0)

varia_thresh.fit(X)

constant_columns = [column for column in X.columns if column not in X.columns[varia_thresh.get_support()]]
```

```python
constant_columns
```

```python
import pickle
```

```python
def model_training(regression_type):
    """
    Model training function
    """

    try:
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
        if regression_type == "linear":
            linear = LinearRegression()
            model = linear.fit(X_train, y_train)
            y_pred = model.predict(X_test)
            score = r2_score(y_test, y_pred)
            if score >= 0.70:
                pickle.dump(model, open("Linear.pkl", "wb"))
            return f"The average score of the model is {score}"
        else:
            alpha = [10, 0.001, 50, 200]
            for i in alpha:
                ridge = Ridge(alpha=i)
                model = ridge.fit(X_train, y_train)
                y_pred = model.predict(X_test)
                score = r2_score(y_test, y_pred)
                if score >= 0.70:
                    pickle.dump(model, open("Ridge.pkl", "wb"))
                return f"The average score of the model is {score}"
    except Exception as e:
        return f"Error: {e}"
```

```python
model_training("linear")
```
